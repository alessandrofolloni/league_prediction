```python
df['Gls'] = [66,53,36,37,53,31,71,56,60,33,64,48,77,50,48,24,
             47,31,42,47]
df['Gls casa'] = [38,29,15,22,30,20,37,38,32,18,41,
                  25,40,28,30,12,24,18,15,30]
df['Gls trasferta'] = df['Gls'] - df['Gls casa']
```

```python
df['Gls conceded']= [48,49,69,49,43,59,42,33,30,46,43,52,28,
                       38,62,71,61,62,41,48]
df['Gls conc casa'] = [26,14,31,23,22,32,14,15,18,24,20,22,15,15,
                       26,28,23,30,19,20]
df['Gls conc trasf'] = df['Gls conceded'] - df['Gls conc casa']
```

```python
#Normalization required for Poisson
df['Gls'] = df['Gls']/38
df['Gls conceded'] = df['Gls conceded']/38
df['Gls casa'] = df['Gls casa']/19
df['Gls trasferta'] = df['Gls trasferta']/19
df['Gls conc casa'] = df['Gls conc casa']/19
df['Gls conc trasf'] = df['Gls conc trasf']/19
df['Points 23/24'] = 0
df['W'] = 0
df['D'] = 0
df['L'] = 0
```

```python
df.head(20)
```

```python
from itertools import combinations

elementi = list(range(20))  # numero giornate

combinazioni = list(combinations(elementi, 2))

# Aggiungi combinazioni invertite
for coppia in list(combinazioni):
    combinazioni.append((coppia[1], coppia[0]))

combinazioni.sort()
len(combinazioni)
```

```python
import time
tempo_inizio = time.time()
```

```python
def print_score(home,guest,goal1,goal2):
  print(f'{home} vs {guest} had the score: {goal1} - {goal2}')
```

```python
%%capture
iterations = 1000

elenco_dataframe = []
for i in range(iterations):
    nome_dataframe = f'df_{i}'
    df_copia = df.copy()
    elenco_dataframe.append(df_copia)
    # Create variables DataFrame df_0 ... df_999
    vars()[nome_dataframe] = df_copia

for n in range(iterations):
  for i in combinazioni:
    #access to i[0] and i[1]
    idx1 = df['Gls casa'].iloc[i[0]] * df['Gls conc trasf'].iloc[i[1]]
    idx2 = df['Gls trasferta'].iloc[i[1]] * df['Gls conc casa'].iloc[i[0]]
    s_c = np.random.poisson(idx1)
    s_g = np.random.poisson(idx2)

    # -------- uncomment to print the scores ---------
    #print_score(df['Squad'].iloc[i[0]],df['Squad'].iloc[i[1]],s_c,s_g)

    if s_c > s_g :
      elenco_dataframe[n]['Points 23/24'].iloc[i[0]] += 3
      elenco_dataframe[n]['W'].iloc[i[0]] += 1
      elenco_dataframe[n]['L'].iloc[i[1]] += 1
    elif s_c == s_g :
      elenco_dataframe[n]['Points 23/24'].iloc[i[0]] += 1
      elenco_dataframe[n]['Points 23/24'].iloc[i[1]] += 1
      elenco_dataframe[n]['D'].iloc[i[0]] += 1
      elenco_dataframe[n]['D'].iloc[i[1]] += 1
    else:
      elenco_dataframe[n]['Points 23/24'].iloc[i[1]] += 3
      elenco_dataframe[n]['W'].iloc[i[1]] += 1
      elenco_dataframe[n]['L'].iloc[i[0]] += 1
```

```python
tempo_fine = time.time()
tempo_trascorso = tempo_fine - tempo_inizio
tempo_trascorso = tempo_trascorso/60

print(f"Tempo impiegato: {tempo_trascorso} minuti")
```
